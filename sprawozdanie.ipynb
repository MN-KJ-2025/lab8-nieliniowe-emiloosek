Dla funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć** znajdź miejsca zerowe przy użyciu:

### **Zadanie 6.**

Korzystając z przedziałów wyznaczonych w **Zadaniu 1.** znajdź miejsca zerowe funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć**, przy użyciu:

- metody bisekcji,
- metody siecznych,
- metody Newtona,

z tolerancją równą $1\mathrm{e}{-10}$.

Zbadaj dokładność (względem rozwiązania z **Zadania 5.**) i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach. 

```python
# ===== Twoja implementacja tutaj =====


przedzial1 = (-0.2,0.1)
przedzial2= (0.5,1)
tol = 1e-10

print("Pierwsze miejsce zerowe")
bisection = main.bisection(przedzial1[0],przedzial1[1],main.func,tol,100)
print("Bisekcja: ", bisection)
print("błąd: ", abs(root1mz.x - bisection[0]) )
secant = main.secant(przedzial1[0],przedzial1[1],main.func,tol,100)
print("Sieczne: ", secant )
print("błąd: ", abs(root1mz.x - secant[0]) )
newton = main.newton(main.func,main.dfunc,main.ddfunc,przedzial1[0],przedzial1[1],tol,100)
print("Newtona: ", newton )
print("błąd: ", abs(root1mz.x - newton[0]),"\n")
print("Drugie miejsce zerowe")
bisection2 = main.bisection(przedzial2[0],przedzial2[1],main.func,tol,100)
print("Bisekcja: ", bisection2 )
print("błąd: ", abs(root2mz.x - bisection2[0]) )
secant2 = main.secant(przedzial2[0],przedzial2[1],main.func,tol,100)
print("Sieczne: ", secant2)
print("błąd: ", abs(root2mz.x - secant2[0]) )
newton2 = main.newton(main.func,main.dfunc,main.ddfunc,przedzial2[0],przedzial2[1],tol,100)
print("Newtona: ", newton2)
print("błąd: ", abs(root2mz.x - newton2[0]),"\n")



liczbaiter = np.arange(1,100,2)
plt.figure()
plt.subplot(2,1,1)
plt.title("Wykres czasu wykonania funkcji dla pierwszego miejsca zerowego")
plt.xlabel('liczba iteracji')
plt.ylabel('czas wykonania')
wyniki1=[[],[],[]]
wyniki2=[[],[],[]]
for it in liczbaiter:
    time = %timeit -r 5 -n 10 -q -o main.bisection(przedzial1[0],przedzial1[1],main.func,tol,it)
    wyniki1[0].append(time.average)
    time = %timeit -r 5 -n 10 -q -o main.secant(przedzial1[0],przedzial1[1],main.func,tol,it)
    wyniki1[1].append(time.average)
    time = %timeit -r 5 -n 10 -q -o main.newton(main.func,main.dfunc,main.ddfunc,przedzial1[0],przedzial1[1],tol,it)
    wyniki1[2].append(time.average)
plt.scatter(liczbaiter,wyniki1[0], label = "bisekcja")
plt.scatter(liczbaiter,wyniki1[1], label = "metoda siecznych")
plt.scatter(liczbaiter,wyniki1[2], label = "metoda newtona")
plt.legend()
plt.show()
plt.subplot(2,1,2)
plt.title("Wykres czasu wykonania funkcji dla drugiego miejsca zerowego")
plt.xlabel('liczba iteracji')
plt.ylabel('czas wykonania')
for it in liczbaiter:
    time = %timeit -r 5 -n 10 -q -o main.bisection(przedzial2[0],przedzial2[1],main.func,tol,it)
    wyniki2[0].append(time.average)
    time = %timeit -r 5 -n 10 -q -o main.secant(przedzial2[0],przedzial2[1],main.func,tol,it)
    wyniki2[1].append(time.average)
    time = %timeit -r 5 -n 10 -q -o main.newton(main.func,main.dfunc,main.ddfunc,przedzial2[0],przedzial2[1],tol,it)
    wyniki2[2].append(time.average)
plt.scatter(liczbaiter,wyniki2[0], label = "bisekcja")
plt.scatter(liczbaiter,wyniki2[1], label = "metoda siecznych")
plt.scatter(liczbaiter,wyniki2[2], label = "metoda newtona")
plt.legend()
plt.show()
```
